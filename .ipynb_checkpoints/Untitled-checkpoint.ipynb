```python
import matplotlib.pyplot as plt
import requests
import pandas as pd
import json
from citipy import citipy 
from config import Weather_KEY
import random
```

```python
from citipy import citipy 
cities = []
while len(cities)<850:
     x = random.randint(-90,90)
     y = random.randint(-180, 180)
     city = citipy.nearest_city(x,y).city_name
     if city not in cities: 
         cities.append(city)
```

```python
len(cities)
```

```python
Temp=[]
country=[]
long=[]
lat=[]
wind_speed=[]
Humidity=[]
Cloudiness =[]
city_weather=[]
```

```python
x = 0
for city in cities :
    print(x)
    try:
        respones = requests.get (f'https://api.openweathermap.org/data/2.5/weather?q={city}&units=imperial&appid={Weather_KEY}').json()
        Temp.append(respones["main"]["temp"])
        country.append(respones["sys"]["country"]) 
        long.append(respones["coord"]["lon"])
        lat.append(respones["coord"]["lat"])
        wind_speed.append(respones["wind"]["speed"])
        Humidity.append(respones["main"]["humidity"])
        Cloudiness.append(respones["clouds"]["all"])
        city_weather.append(city)
        print(f'{city} added')
    except:
        print(f'{city} causing an error')
    x += 1    

```

```python
Data_df=pd.DataFrame({
    "city":city_weather,
    "country":country,
    "long":long,
    "lat":lat,
    "Temperature":Temp,
    "Humidity (%)":Humidity,
    "Cloudiness(%)":Cloudiness,
    "Wind Speed (mph)":wind_speed    
})
Data_df
```

```python
plt.scatter (Data_df['lat'],Data_df['Temperature'])
plt.xlabel('lat')
plt.ylabel('Temperature')
plt.title("Temperature (F) vs. Latitude")
plt.show()
```

```python
plt.scatter (Data_df['lat'],Data_df['Humidity (%)'])
plt.xlabel('lat')
plt.ylabel('Humidity (%)')
plt.title("Humidity (%) vs. Latitude")
plt.show()
```

```python
plt.scatter (Data_df['lat'],Data_df['Cloudiness(%)'])
plt.xlabel('lat')
plt.ylabel('Cloudiness(%)')
plt.title(" Cloudiness(%)vs. Latitude")
plt.show()
```

```python
plt.scatter (Data_df['lat'],Data_df['Wind Speed (mph)'])
plt.xlabel('lat')
plt.ylabel('Wind Speed (mph)')
plt.title("Wind Speed (mph) vs. Latitude")
plt.show()
```
